```python
# import pandas as pd
import pandas as pd

# import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

#read csv file
df = pd.read_csv('index.csv', encoding= 'latin1')
print(df.head())
```

```python
# checking datatypes
df['money'] = df['money'].astype(float)
print(df['money'])
```

```python
# checking for missing values
df.fillna(0)

#dropping duplicate values
df.drop_duplicates()

print(df)
```

```python
# basic statistics of numerical column
print(df.describe())
```

```python
# groupings on a categorical column and compute the mean of a numerical column for each group
grouped = df.groupby('coffee_name').agg({'money': 'mean'})
print(grouped)
```

```python
# ploting a time series graph with coffee_name vs date
# Plotting Age vs. Name with plt
plt.plot( df['date'], df['coffee_name'])
plt.xlim(-5, 50)  # Expand x-axis range
plt.ylim(-2, 15)   # Expand y-axis range
plt.xlabel('Date')
plt.ylabel('Coffee')
plt.title('Coffee Sales Over The Year')
plt.show()
```

```python
# Plotting a bar chart for coffee_name by money
df.groupby('coffee_name')['money'].mean().plot(kind='bar')
plt.xlabel('Coffee')
plt.ylabel('Average Amount Sold')
plt.title('Average Coffe Sold')
plt.show()
```

```python
# Plotting a histogram for Age
df['money'].plot(kind='hist', bins=10)
plt.xlabel('Money')
plt.title('Money Distribution')
plt.show()
```

```python
# Plotting a scatter plot of Age vs. State
df.plot(kind='scatter', x='coffee_name', y='money')
plt.title('Coffee vs Money')
plt.show()
```
