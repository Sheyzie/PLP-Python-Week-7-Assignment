```python
# lets import panda as pd and matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

# creating data dictionary for analysis
data = {
    'Name': ['Pascal', 'Ayotunde', 'Victor', 'Sandra', 'Deborah', 'Gift', 'Favor'],
    'Age': [29, 30, 27, 26, 25, 26, 25],
    'State': ['Imo', 'Ondo', 'Anambra', 'Edo', 'Edo', 'Cros River', 'Imo']
}

#creating dataframe with pandas
df = pd.DataFrame(data)
print(df)
```

```python
# Plotting Age vs. Name with plt
plt.plot(df['Name'], df['Age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Name vs Age')
plt.show()
```

```python
# Plotting a bar chart for Age by State
df.groupby('State')['Age'].mean().plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Average Age')
plt.title('Average Age by State')
plt.show()
```

```python
# Plotting a histogram for Age
df['Age'].plot(kind='hist', bins=10)
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()
```

```python
# Plotting a scatter plot of Age vs. State
df.plot(kind='scatter', x='State', y='Age')
plt.title('State vs Age')
plt.show()
```
